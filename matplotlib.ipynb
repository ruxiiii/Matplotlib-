```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1,2,3]
y = [4,5,6]
plt.figure(figsize=(5,5))

#first line red

#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')
#plt.plot(x,y, label = "2x", color = "red",linewidth=2,linestyle ="--",markersize=20,markeredgecolor = "blue")
plt.plot(x,y,"r--",label="2x")

#second line blue 
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[2:6], x2[2:6]**2, 'b^-', label='X^2')
#plt.figure(figsize=(8,5),dpi=100)

#add a legend
plt.legend()

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([1,2,3])

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

#show plot
plt.show()
```

Bar Graph

```python
labels = ["A","B","C"]
values = [1,4,2]

plt.figure(figsize=(5,5),dpi=100)

patterns = ["o","*","/"]
bars = plt.bar(labels,values)

for bar in bars:
  bar.set_hatch(patterns.pop(0))
plt.show()
plt.savefig('barchart.png', dpi=300)
```

Petrol Prices

```python
df = pd.read_csv("petrol-2.csv")
df.head()
```

```python
#size of figure
plt.Figure(figsize=(5,5))

#title
plt.title("Petrol Prices over Time", fontdict={'fontweight':'bold', 'fontsize': 18})


plt.plot(df.Year,df.Japan,"b.-",label = "Japan")
plt.plot(df.Year,df.Australia,"y.-",label = "Australia")
plt.plot(df.Year,df["South Korea"],"g.-",label = "South Korea")
plt.plot(df.Year,df.USA,"r.-",label = "USA")

#making year jump by 3
plt.xticks(df.Year[::3])

#labelling
plt.xlabel("Year")
plt.ylabel("US Dollars")

plt.legend()
plt.savefig("Petrol_price_figure.png",dpi=100)
plt.show()
```


###Loading the Fifa Dataset

```python
fifa = pd.read_csv("fifa.csv")
fifa.head()
```


Histogram

```python

#The towers or bars of a histogram are called bins
bins = [20,50,60,70,80,90,100]
#bins=[90,100] would tell that there are player betwween them as well


#size of fig
plt.figure(figsize=(8,5))
plt.hist(fifa.Overall,bins=bins,color="blue")

plt.xticks(bins)
plt.xlabel("Overall Score")
plt.ylabel("Number of Players")

plt.savefig("fifa_histogram.png")

plt.show()
```

Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='yellow', linewidth=2)
    # change fill color
    box.set(facecolor = 'pink' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
